```python
import pandas as pd
```

```python
generated_output = 'The system cannot answer this question because the available data only covers from 2024-12-01 to 2025-05-20, and the requested data is for 2023.'
refrence_output = "The system cannot answer this question because the data does not contain that information."
```

```python
from sentence_transformers import SentenceTransformer, util

# Load pre-trained SBERT model
model = SentenceTransformer("all-MiniLM-L6-v2")


# Compute embeddings
embedding1 = model.encode(generated_output, convert_to_tensor=True)
embedding2 = model.encode(refrence_output, convert_to_tensor=True)

# Compute cosine similarity
cosine_score = util.pytorch_cos_sim(embedding1, embedding2)

print(f"Cosine Similarity: {cosine_score.item():.4f}")
```

```python
def compute_bert_score(cands, refs, lang="en", idf=False, device="cpu", verbose=True):
    from bert_score import score
    P, R, F1 = score(
        cands,
        refs,
        lang=lang,
        verbose=verbose,
        model_type="bert-base-uncased",
    )
    return P, R, F1
```

```python
benchmark_run=pd.read_csv("../../scripts/benchmarks/benchmark_results_lnGxEDxx.csv")
```

```python
# cands = [generated_output.lower()]
# refs = [refrence_output.lower()]
cands = benchmark_run["generated_answer"].tolist()
refs = benchmark_run["reference_answer"].tolist()
```

```python
P, R, F1 = compute_bert_score(cands, refs,verbose=False,idf=True)
print(f"F1: {F1.mean().item()}")
print(f"Precision: {P.mean().item()}")
print(f"Recall: {R.mean().item()}")
```
