```python
import base64
from IPython.display import Image, display
```

```python
def display_plot(image_base64_str):
    image_data = base64.b64decode(image_base64_str.split(",")[1])
    display(Image(data=image_data))
```

```python
# 1. Line plot: Sales over years
img1 = data_vizualization(
    "line", "Year", "Sales", [2019, 2020, 2021, 2022], [100, 150, 200, 250]
)
display_plot(img1)

# 2. Bar chart: Product quantities
img2 = data_vizualization("bar", "Product", "Quantity", ["A", "B", "C"], [10, 20, 15])
display_plot(img2)

# 3. Scatter plot: Age vs Score
img3 = data_vizualization("scatter", "Age", "Score", [18, 22, 30, 40], [70, 80, 65, 90])
display_plot(img3)

# 4. Pie chart: Budget by Department
img4 = data_vizualization(
    "pie", "Department", "Budget Share", ["HR", "Eng", "Mkt", "Sales"], [25, 40, 20, 15]
)
display_plot(img4)

# 5. Line plot: Temperature change
img5 = data_vizualization(
    "line", "Month", "Temperature", ["Jan", "Feb", "Mar", "Apr"], [30, 35, 40, 38]
)
display_plot(img5)

# 6. Bar chart: Revenue by Region
img6 = data_vizualization(
    "bar",
    "Region",
    "Revenue",
    ["North", "South", "East", "West"],
    [1200, 950, 1100, 1020],
)
display_plot(img6)

# 7. Scatter plot: Study Hours vs Exam Score
img7 = data_vizualization(
    "scatter", "Hours Studied", "Exam Score", [1, 2, 3, 4, 5], [50, 60, 70, 80, 90]
)
display_plot(img7)

# 8. Pie chart: Market Share
img8 = data_vizualization(
    "pie",
    "Company",
    "Market Share",
    ["Apple", "Samsung", "Xiaomi", "Others"],
    [40, 30, 20, 10],
)
display_plot(img8)

# 9. Line plot: Website Visitors
img9 = data_vizualization(
    "line",
    "Day",
    "Visitors",
    ["Mon", "Tue", "Wed", "Thu", "Fri"],
    [300, 400, 350, 500, 450],
)
display_plot(img9)

# 10. Bar chart: Task Completion
img10 = data_vizualization(
    "bar", "Task", "Completion %", ["T1", "T2", "T3", "T4"], [90, 75, 60, 85]
)
display_plot(img10)
```
